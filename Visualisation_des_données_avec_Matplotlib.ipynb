**Etape 2 : Initiation à l'analyse statistique**

```python
#21
# Tracer un histogramme des notes (grade)
plt.hist(data_cleaned['Grade'], bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Distribution of Grades')

# Afficher le graphique
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#22
from scipy import stats

# Calculer les statistiques descriptives
mean_grade = data_cleaned['Grade'].mean()
median_grade = data_cleaned['Grade'].median()
mode_result = stats.mode(data_cleaned['Grade'])  # Calculate the mode
if mode_result.mode.size > 0:  # Check if mode exists
    mode_grade = mode_result.mode # Extract the mode value if it exists
else:
    mode_grade = None  # Set mode_grade to None if no mode exists
min_grade = data_cleaned['Grade'].min()
max_grade = data_cleaned['Grade'].max()

# Tracer un histogramme des notes (grade)
plt.hist(data_cleaned['Grade'], bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Distribution of Grades')

# Afficher les statistiques sur l'histogramme
plt.axvline(mean_grade, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean_grade:.2f}')
plt.axvline(median_grade, color='g', linestyle='dashed', linewidth=1, label=f'Median: {median_grade}')
plt.axvline(mode_grade, color='b', linestyle='dashed', linewidth=1, label=f'Mode: {mode_grade}')
plt.axvline(min_grade, color='k', linestyle='dashed', linewidth=1, label=f'Min: {min_grade}')
plt.axvline(max_grade, color='m', linestyle='dashed', linewidth=1, label=f'Max: {max_grade}')

# Ajouter une légende
plt.legend()

# Afficher le graphique
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```

```python
#23
# Calculer les statistiques descriptives
min_val = data_cleaned['Grade'].min()
max_val = data_cleaned['Grade'].max()
mean_val = data_cleaned['Grade'].mean()
median_val = data_cleaned['Grade'].median()
mode_val = stats.mode(data_cleaned['Grade'])  # Calculate the mode
if mode_val.mode.size > 0:  # Check if mode exists
    mode_grade = mode_val.mode # Extract the mode value if it exists
else:
    mode_grade = None  # Set mode_grade to None if no mode exists

# Tracer un histogramme des notes (grade)
plt.figure(figsize=(8, 5))  # Définir la taille de la figure

plt.hist(data_cleaned['Grade'], bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Distribution of Grades')

# Ajouter des lignes pour chaque statistique calculée
plt.axvline(x=min_val, color='gray', linestyle='dashed', linewidth=2, label=f'Min: {min_val}')
plt.axvline(x=max_val, color='gray', linestyle='dashed', linewidth=2, label=f'Max: {max_val}')
plt.axvline(x=mean_val, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(x=median_val, color='g', linestyle='dashed', linewidth=2, label=f'Median: {median_val}')
# Check if mode exists before plotting
if mode_grade is not None:
    plt.axvline(x=mode_grade, color='b', linestyle='dashed', linewidth=2, label=f'Mode: {mode_grade}')


# Ajouter une légende
plt.legend()

# Afficher le graphique
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#24
# Calculer les statistiques descriptives
q1_val = pd.Series(data_cleaned['Grade']).quantile(0.25)
q3_val = pd.Series(data_cleaned['Grade']).quantile(0.75)

# Tracer un boxplot des notes (grade)
plt.figure(figsize=(8, 5))
plt.boxplot(data_cleaned['Grade'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel('Values')
plt.title('Boxplot of Grades')

# Ajouter des annotations pour les valeurs min, max, médianes et quartiles
plt.axvline(x=min_val, color='gray', linestyle='dashed', linewidth=2)
plt.axvline(x=max_val, color='gray', linestyle='dashed', linewidth=2)
plt.axvline(x=median_val, color='red', linestyle='dashed', linewidth=2)
plt.axvline(x=q1_val, color='green', linestyle='dashed', linewidth=2)
plt.axvline(x=q3_val, color='green', linestyle='dashed', linewidth=2)

# Ajouter des annotations textuelles pour les valeurs
plt.text(min_val, 1.1, f'Min: {min_val}', verticalalignment='center', color='blue')
plt.text(max_val, 1.1, f'Max: {max_val}', verticalalignment='center', color='blue')
plt.text(median_val, 1.1, f'Median: {median_val}', verticalalignment='center', color='red')
plt.text(q1_val, 1.1, f'Q1: {q1_val}', verticalalignment='center', color='green')
plt.text(q3_val, 1.1, f'Q3: {q3_val}', verticalalignment='center', color='green')

# Afficher le graphique
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
